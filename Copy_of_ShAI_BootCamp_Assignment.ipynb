```python
# Task 1

# no of rows
print(f"No of rows: {len(df)}") # df.shape[0]
print()
# no of columns
print(f"No of rows: {len(df.columns)}") # df.shape[1]
print()
print(f"Dtypes for each column: {df.info()}")
print()
print(f"No of nan in each col: {df.isnull().sum()}")
```

```python
# Task 2

print(f"Descriptive Statistics: \n{df.describe()}")

print()

print(f"Min of Salaries: {df['TotalPay'].min()}")
print(f"Max of Salaries: {df['TotalPay'].max()}")
print(f"Range of Salaries: {df['TotalPay'].max() - df['TotalPay'].min()}")

```

```python
# Task 3

# Deletion is used
# The reason is that there are a lot of nan values in Benefits and Status
# Then removing the rest rows which contain nan

# df.dropna(axis=1).head()

dfClean = df.drop(['Notes',"Benefits","Status"], axis=1)
dfClean = dfClean.dropna(axis=0)
dfClean.isnull().sum()
df = dfClean
```

```python
# Task 4

import matplotlib.pyplot as plt

plt.hist( df["TotalPay"].values, edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()

value_counts = dfClean['Year'].value_counts()

plt.pie(value_counts, labels=['2011','2012','2013','2014'])

plt.show()
```

```python
# Task 5
print(df.groupby(['Year']).head())
print("ok")
print(f"Grouping with year: {df.groupby(['Year'])['TotalPay'].describe()}")
print(f"Grouping with job title: {df.groupby(['JobTitle'])['TotalPay'].describe()}")
```

```python
# Task 6
print("Correlation between Year and Salay:")
print(df['TotalPay'].corr(df['Year']))
plt.scatter(df["Year"], df['TotalPay'])
plt.title(f'Salary vs. Year')
plt.show()

print("Correlation between OtherPay and Salay:")
print(df['TotalPay'].corr(df['OtherPay']))
plt.scatter(df["OtherPay"], df['TotalPay'])
plt.title(f'Salary vs. OtherPay')
plt.show()

print("Correlation between BasePay and Salay:")
print(df['TotalPay'].corr(df['BasePay']))
plt.scatter(df["BasePay"], df['TotalPay'])
plt.title(f'Salary vs. BasePay')
plt.show()

print("Correlation between OvertimePay and Salay:")
print(df['TotalPay'].corr(df['OvertimePay']))
plt.scatter(df["OvertimePay"], df['TotalPay'])
plt.title(f'Salary vs. OvertimePay')
plt.show()

print("Correlation between No of persons in a Job and Salay:")
print(df['TotalPay'].corr(pd.Series(df['JobTitle'].value_counts().values)))
plt.scatter(df['JobTitle'], df['TotalPay'])
plt.title(f'Salary vs. Job Count')
plt.show()
```

```python
# Task 6

# very weak positive correlation between year and salary

# moderate positive correlation between other pay and salary

# strong positive correlation between base pay and salary

# moderate positive correlation between overtime pay and salary

# strong positive correlation between the number of persons in a job and salary

# In the end, focus on the base pay and no of people in the current job -> which makes sense
```
